from torch.nn.modules.linear import Linear
from torch.nn import Sequential
from torchtune.modules.peft import LoRALinear
def replace_qlora(module, name="<ROOT>"):
    '''
    Recursively put qlora in nn.linear.

    set module = net to start code.
    '''
    # go through all attributes of module nn.module (e.g. network or layer) and put batch norms if present
    for attr_str, _ in module.named_children():
        target_attr = getattr(module, attr_str)
        if type(target_attr) == Linear and "lora" not in attr_str:
            print('replacing: ', name, attr_str)
            try:
                new_ql = Sequential(LoRALinear(in_dim=target_attr.in_features, out_dim=target_attr.out_features, rank=min([target_attr.in_features, target_attr.out_features, 4]), alpha=0.1, quantize_base=True), 
                           BiasLayer(dim=target_attr.out_features))

                setattr(module, attr_str, new_ql)
            except:
                print("failed, check dims: ", target_attr.in_features, target_attr.out_features)

    # iterate through immediate child modules. Note, the recursion is done by our code no need to use named_modules()
    for name, immediate_child_module in module.named_children():
        replace_qlora(immediate_child_module, name)

replace_qlora(model)

```python
import torch
class BiasLayer(torch.nn.Module):
    def __init__(self, dim) -> None:
        super().__init__()
        bias_value = torch.randn((dim))
        self.bias_layer = torch.nn.Parameter(bias_value)
    
    def forward(self, x):
        return x + self.bias_layer
```
